# Data Visualization

```python
len(os.listdir(data_dir))
```

```python
path = []
for classes in os.listdir(data_dir):
    path.append(data_dir+'/'+classes)
```

```python
images_path = []
for i, img_path in enumerate(path):
    images_path.append(img_path+'/'+random.choice(os.listdir(path[i])))
    
## Plot Images
plt.figure()
plt.subplots(3, 4)
for i, images in enumerate(images_path):
    plt.subplot(3, 4, 1+i)
    image = plt.imread(images)
    plt.imshow(image)
    plt.axis(False)
```
